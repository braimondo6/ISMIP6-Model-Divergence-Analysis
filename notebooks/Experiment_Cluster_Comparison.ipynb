```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from matplotlib import cm # For colormaps
import itertools # For markers
from skimage.transform import resize # For resize
import numpy as np
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler
from matplotlib.patches import Ellipse # Import Ellipse for 2D
import scipy.stats as stats # For chi-squared for confidence ellipse scaling
import xarray as xr # For dummy data
import datetime # For dummy data
import seaborn as sns # For heatmap
from sklearn.metrics import adjusted_rand_score # For cluster similarity
```

```python
# ---- USE THIS CODE ONLY FOR CORE EXPERIMENTS ----
# ---- Or any experiment with same number of participating models ----

# --- USER CONFIGURATION ---
# Define the actual experiment IDs for your two experiments.
# These IDs will be used in folder and file names for clarity.
expx_id = "expa02" # <--- Set the ID for your first experiment
expy_id = "expa03" # <--- Set the ID for your second experiment

# Define the paths to your experiment folders containing NetCDF files
expx_folder_path = r"C:\Users\benra\lithk_expa02" # <--- IMPORTANT: SET YOUR PATH HERE
expy_folder_path = r"C:\Users\benra\lithk_expa03" # <--- IMPORTANT: SET YOUR PATH HERE

# Define the variable name exactly as it appears in your NetCDF files
variable_name_in_netcdf = "lithk" # <--- IMPORTANT: SET YOUR VARIABLE NAME HERE

# PCA and K-Means parameters
n_components = 4 # Number of PCA components
n_clusters = 5 # Number of K-Means clusters

# Define time windows (decades as suggested)
time_windows = {
    '2015-2025': (0, 10),   # indices for 2015-2025
    '2015-2035': (0, 20),   # indices for 2015-2035  
    '2015-2045': (0, 30),   # indices for 2015-2045
    '2015-2055': (0, 40),   # indices for 2015-2055
    '2015-2065': (0, 50),   # indices for 2015-2065
    '2015-2075': (0, 60),   # indices for 2015-2075
    '2015-2085': (0, 70),   # indices for 2015-2085
    '2015-2100': (0, 85)    # full time series
}
# --- END USER CONFIGURATION ---


# --- Function to load NetCDF files from a folder ---
def load_netcdf_files_from_folder(folder_path, var_name_in_nc):
    """
    Loads NetCDF files from a specified folder, extracts a specific variable,
    and returns a list of xarray DataArrays and corresponding model names.
    """
    var_list_loaded = []
    model_names_loaded = []

    if not os.path.isdir(folder_path):
        print(f"Error: Folder not found at '{folder_path}'. Please check the path.")
        return [], []

    # Iterate through files in the folder (sorted for consistent model order)
    for filename in sorted(os.listdir(folder_path)):
        if filename.endswith(".nc"):
            filepath = os.path.join(folder_path, filename)
            try:
                with xr.open_dataset(filepath) as ds:
                    if var_name_in_nc in ds:
                        var_list_loaded.append(ds[var_name_in_nc])
                        
                        # --- MODIFIED LINE HERE ---
                        # Extract the base filename without extension
                        filename_base = os.path.splitext(filename)[0]
                        # Remove the '_expX' suffix (e.g., '_exp05' or '_exp06')
                        # and then remove the 'lithk_GIS_' prefix to get the clean model name.
                        parts = filename_base.rsplit('_', 1) # Splits at the last underscore
                        if len(parts) > 1: # Ensure there was an underscore to split by
                            model_name_with_prefix = parts[0]
                            clean_model_name = model_name_with_prefix.replace("lithk_GIS_", "")
                            model_names_loaded.append(clean_model_name)
                        else:
                            # Fallback if filename format is unexpected (e.g., no _expid)
                            model_names_loaded.append(filename_base.replace("lithk_GIS_", ""))
                        # --- END MODIFIED LINE ---
                        
                    else:
                        print(f"Warning: Variable '{var_name_in_nc}' not found in '{filename}'. Skipping.")
            except Exception as e:
                print(f"Error loading '{filepath}': {e}")
    return var_list_loaded, model_names_loaded

# Initialize a dictionary to store all clustering results.
# This will hold cluster labels for each experiment and time window.
all_clustering_results = {}

# Map actual experiment IDs to their folder paths
experiment_configs = {
    expx_id: expx_folder_path,
    expy_id: expy_folder_path
}

# --- Main Processing Loop for Each Experiment ---
for current_exp_id, folder_path in experiment_configs.items():
    print(f"\n--- Processing Experiment: {current_exp_id} from {folder_path} ---")

    # Load data for the current experiment
    var_list, participating_models = load_netcdf_files_from_folder(
        folder_path, variable_name_in_netcdf
    )

    if not var_list:
        print(f"No valid NetCDF files found or variable '{variable_name_in_netcdf}' missing in '{folder_path}'. Skipping {current_exp_id}.")
        continue

    # Set the variable name for plotting purposes
    var_name = variable_name_in_netcdf

    # Define base output directory for the current experiment's plots
    base_output_dir = os.path.join("Research_Plots", f"{current_exp_id}_{var_name}_plots")
    os.makedirs(base_output_dir, exist_ok=True)

    # Loop over all defined time windows
    for time_window_label, (start_idx, end_idx) in time_windows.items():
        print(f"\n Processing: {time_window_label} for {current_exp_id}...")
        
        # Create directory for each time window's plots
        folder_name = time_window_label.replace(" ", "_").replace("(","").replace(")", "")
        time_window_output_dir = os.path.join(base_output_dir, folder_name)
        os.makedirs(time_window_output_dir, exist_ok=True)

        # Get grid coordinates from the first model in var_list
        # This assumes all models share the same grid structure.
        first_var_dataarray = var_list[0]
        x_coords = first_var_dataarray['x'].values if 'x' in first_var_dataarray.coords else first_var_dataarray.coords['x'].values
        y_coords = first_var_dataarray['y'].values if 'y' in first_var_dataarray.coords else first_var_dataarray.coords['y'].values
        
        rows = len(y_coords)
        cols = len(x_coords)
        matrix_size = rows * cols

        # Step 1: Compute variable deltas for the current time window and remove NaNs
        current_window_deltas = []
        
        for var_model_data in var_list:
            # Calculate delta for the current time window using index-based selection
            delta = var_model_data.isel(time=end_idx) - var_model_data.isel(time=start_idx)
            
            # Ensure delta is resized to the common grid shape if necessary, then flatten
            delta_array = delta.values if hasattr(delta, 'values') else delta
            resized_delta = resize(delta_array, (rows, cols), anti_aliasing=True)
            flat_delta = resized_delta.flatten()
            
            current_window_deltas.append(flat_delta)

        all_deltas = np.stack(current_window_deltas)
        mask = ~np.any(np.isnan(all_deltas), axis=0) # Mask out grid points with NaN in any model
        data_matrix = all_deltas[:, mask]  # Final clean matrix (n_models, n_valid_points)
        
        # Get valid grid indices based on the mask
        gd = np.where(mask)[0]

        # Step 2: PCA (describes variability across models for this time window)
        scaler = StandardScaler()
        data_matrix_scaled = scaler.fit_transform(data_matrix)
        
        pca = PCA(n_components=n_components)
        pca_scores = pca.fit_transform(data_matrix_scaled)
        
        correct_pca_components = pca.components_ # Principal components (eigenvectors)

        # Step 3: K-Means Clustering
        kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init='auto') 
        labels = kmeans.fit_predict(pca_scores)
        
        # Store the clustering results for later comparison heatmap
        key = (current_exp_id, time_window_label)
        all_clustering_results[key] = {
            'labels': labels,
            'models': participating_models # Store models to ensure consistency during ARI calculation
        }

        # Create a DataFrame to store model names and their cluster labels
        cluster_results_df = pd.DataFrame({
            'Model': participating_models,
            'Cluster_Label': labels
        })

        # --- Cluster Color and Marker Setup for Plots ---
        if n_clusters <= 9:
            cluster_colors = [cm.get_cmap('Set1')(i) for i in range(n_clusters)]
        elif n_clusters <= 12:
            cluster_colors = [cm.get_cmap('Set3')(i) for i in range(n_clusters)]
        else:
            print("Warning: Number of clusters is > 12. Color distinctness may be challenging.")
            cluster_colors = [cm.get_cmap('hsv')(i/n_clusters) for i in range(n_clusters)]

        markers = itertools.cycle(('o', '^', 's', 'D', 'v', 'P', 'X', '*', 'H', '<', '>'))
        model_to_marker = {model: next(markers) for model in participating_models}

        # --- Step 4a: Plot PCA result in 2D (PC1 vs PC2) with Ellipses ---
        fig_2d, ax_2d = plt.subplots(figsize=(10, 8))

        for i, model_name in enumerate(participating_models): 
            cluster_id = labels[i]
            color = cluster_colors[cluster_id]
            marker = model_to_marker[model_name]

            ax_2d.scatter(
                pca_scores[i, 0], pca_scores[i, 1],
                c=[color], marker=marker, s=250, edgecolor='black', alpha=0.8
            )
        
        # Add Ellipses for each Cluster in 2D Plot
        confidence_level = 0.95
        chi2_val = stats.chi2.ppf(confidence_level, df=2) # For 2D (2 degrees of freedom)
        
        for cluster_id in np.unique(labels):
            cluster_points = pca_scores[labels == cluster_id, :2]
            if len(cluster_points) < 2: continue 

            centroid = np.mean(cluster_points, axis=0)
            covariance = np.cov(cluster_points, rowvar=False)

            eigenvalues, eigenvectors = np.linalg.eigh(covariance)
            order = eigenvalues.argsort()[::-1] # Sort for major/minor axes
            eigenvalues = eigenvalues[order]
            eigenvectors = eigenvectors[:, order]

            angle = np.degrees(np.arctan2(*eigenvectors[:, 0][::-1]))
            width = 2 * np.sqrt(eigenvalues[0] * chi2_val)
            height = 2 * np.sqrt(eigenvalues[1] * chi2_val)

            ellipse_color = cluster_colors[cluster_id]
            ellipse = Ellipse(xy=centroid, width=width, height=height, 
                              angle=angle, edgecolor=ellipse_color, fc='None', 
                              lw=2, alpha=0.7, zorder=0)
            ax_2d.add_patch(ellipse)
            
        # Create custom legend for clusters and models
        cluster_legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=cluster_colors[i], markersize=10, label=f'Cluster {i}') for i in range(n_clusters)]
        model_legend_handles = [plt.Line2D([0], [0], marker=model_to_marker[model_name], color='w', markerfacecolor=cluster_colors[labels[i]], markersize=10, label=model_name, markeredgecolor='black') for i, model_name in enumerate(participating_models)]

        first_legend = ax_2d.legend(handles=cluster_legend_handles, loc='upper left', title='Clusters', bbox_to_anchor=(1.05, 1))
        ax_2d.add_artist(first_legend)
        ax_2d.legend(handles=model_legend_handles, loc='lower left', title='Models', bbox_to_anchor=(1.05, 0))

        ax_2d.set_xlabel("PC1")
        ax_2d.set_ylabel("PC2")
        ax_2d.set_title(f"2D PCA + KMeans Clustering of {var_name} ({time_window_label})")
        plt.tight_layout(rect=[0, 0, 0.85, 1])
        plt.close()

        # --- Step 4b: Plot PCA result in 3D with Transparent Centroid Markers ---
        fig_3d = plt.figure(figsize=(12, 8))
        ax_3d = fig_3d.add_subplot(111, projection='3d')

        for i, model_name in enumerate(participating_models): 
            cluster_id = labels[i]
            color = cluster_colors[cluster_id]
            marker = model_to_marker[model_name]

            ax_3d.scatter(
                pca_scores[i, 0], pca_scores[i, 1], pca_scores[i, 2],
                c=[color], marker=marker, s=150, edgecolor='black', alpha=0.8
            )

        # Add large, transparent markers at cluster centroids in 3D
        for cluster_id in np.unique(labels):
            cluster_points_3d = pca_scores[labels == cluster_id, :]
            if len(cluster_points_3d) > 0:
                centroid_3d = np.mean(cluster_points_3d, axis=0)
                ax_3d.scatter(
                    centroid_3d[0], centroid_3d[1], centroid_3d[2],
                    c=[cluster_colors[cluster_id]], marker='o', s=1000,
                    alpha=0.1, edgecolor=cluster_colors[cluster_id], linewidth=2, zorder=0
                )

        cluster_legend_handles_3d = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=cluster_colors[i], markersize=10, label=f'Cluster {i}') for i in range(n_clusters)]
        model_legend_handles_3d = [plt.Line2D([0], [0], marker=model_to_marker[model_name], color='w', markerfacecolor=cluster_colors[labels[i]], markersize=10, label=model_name, markeredgecolor='black') for i, model_name in enumerate(participating_models)]

        first_legend_3d = ax_3d.legend(handles=cluster_legend_handles_3d, loc='center left', title='Clusters', bbox_to_anchor=(1.05, 0.7))
        ax_3d.add_artist(first_legend_3d)
        ax_3d.legend(handles=model_legend_handles_3d, loc='center left', title='Models', bbox_to_anchor=(1.05, 0.2))

        ax_3d.set_xlabel("PC1")
        ax_3d.set_ylabel("PC2")
        ax_3d.set_zlabel("PC3")
        ax_3d.set_title(f"3D PCA + KMeans Clustering of {var_name} ({time_window_label})")
        plt.tight_layout(rect=[0, 0, 0.85, 1])
        plt.close()
        
        # Step 5: Plot PCs onto the grid
        for pc_idx in range(n_components):
            pc_values = correct_pca_components[pc_idx, :] 
            
            C = np.full(matrix_size, np.nan)
            C[gd.astype(int)] = pc_values
            pc_map = C.reshape((rows, cols)) 
            
            plt.figure(figsize=(12,8))
            vmax = 0.005 # Adjust as needed for your data visualization
            
            im = plt.imshow(pc_map, cmap='bwr_r', origin='lower', vmin=-vmax, vmax=vmax)
            plt.colorbar(im, label=f'PC{pc_idx + 1}')
            plt.xlabel('x')
            plt.ylabel('y')
            plt.title(f'PC #{pc_idx + 1} ({time_window_label}) - Explained Variance: {pca.explained_variance_ratio_[pc_idx]*100:.1f}%')
            plt.axis('equal')
            plt.close()
        
        # Step 6: Plot signed contribution of each PC onto the grid
        for pc_idx in range(n_components):
            pc_values = correct_pca_components[pc_idx, :] 
            
            sum_sq_pc = np.sum(pc_values ** 2)
            if sum_sq_pc == 0:
                signed_contrib = np.zeros_like(pc_values)
            else:
                signed_contrib = 100 * np.sign(pc_values) * (pc_values ** 2) / sum_sq_pc
            
            C = np.full(matrix_size, np.nan)
            C[gd.astype(int)] = signed_contrib
            sc_map = C.reshape((rows, cols))
            
            plt.figure(figsize=(12,8))
            vmax = 0.005 # Adjust as needed for your data visualization
            
            im = plt.imshow(sc_map, cmap='bwr_r', origin='lower', vmin=-vmax, vmax=vmax)
            plt.colorbar(im, label='Signed Contribution %')
            plt.xlabel('x')
            plt.ylabel('y')
            plt.title(f'Grid contribution to PC #{pc_idx + 1} (%) - Explained Variance: {pca.explained_variance_ratio_[pc_idx]*100:.1f}% ({time_window_label})')
            plt.axis('equal')
            plt.close()
        
        # --- Step 7: Plotting the Sum of PC Vectors onto the grid ---
        summed_pc_values = np.sum(correct_pca_components, axis=0)
        ex_var = pca.explained_variance_ratio_
        cumulative_variance_of_first_four = np.cumsum(ex_var)[min(3, n_components -1)] # Get for first 4 or max available
        
        C_summed = np.full(matrix_size, np.nan)
        C_summed[gd.astype(int)] = summed_pc_values
        summed_pc_map = C_summed.reshape((rows, cols))

        plt.figure(figsize=(12,8))
        vmax_summed = np.percentile(np.abs(summed_pc_values[~np.isnan(summed_pc_values)]), 99) * 0.5
        if vmax_summed == 0: vmax_summed = 0.01

        im = plt.imshow(summed_pc_map, cmap='bwr_r', origin='lower', vmin=-vmax_summed, vmax=vmax_summed)
        plt.colorbar(im, label='Sum of All Principal Components')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title(f'Sum of All Principal Components on Grid - Cum. Variance: {cumulative_variance_of_first_four*100:.1f}% ({time_window_label})')
        plt.axis('equal')
        plt.close()

# --- Cluster Similarity Heatmap Generation ---
print("\n--- Generating Cluster Similarity Heatmap ---")

# Ensure a dedicated output directory for comparison plots
comparison_output_dir = os.path.join("Research_Plots", "Cluster_Comparisons")
os.makedirs(comparison_output_dir, exist_ok=True)

# Get sorted keys for consistent ordering in the heatmap (e.g., ('exp05', '2015-2100'), ('exp06', '2015-2100'))
comparison_keys = sorted(list(all_clustering_results.keys()))

if len(comparison_keys) < 2:
    print("Not enough clustering results to generate a similarity heatmap (need at least two experiments/time windows).")
else:
    # Create labels for the heatmap axes (e.g., "exp05_2015-2100")
    heatmap_labels = [f"{exp}_{tw}" for exp, tw in comparison_keys]

    # Initialize similarity matrix
    num_comparisons = len(comparison_keys)
    similarity_matrix = np.zeros((num_comparisons, num_comparisons))

    # Calculate pairwise Adjusted Rand Index (ARI) scores
    for i in range(num_comparisons):
        for j in range(num_comparisons):
            key_i = comparison_keys[i]
            key_j = comparison_keys[j]

            labels_i = all_clustering_results[key_i]['labels']
            models_i = all_clustering_results[key_i]['models']

            labels_j = all_clustering_results[key_j]['labels']
            models_j = all_clustering_results[key_j]['models']

            # IMPORTANT: For ARI to be meaningful, the underlying data points (models)
            # being clustered must be the same and in the same order.
            # If your 'participating_models' lists differ or are not in the same order
            # between compared keys, the ARI will be misleading.
            # You might need to implement logic to find common models and
            # extract their corresponding labels before calculating ARI.
            if not np.array_equal(models_i, models_j):
                print(f"Warning: Model lists differ or are not in the same order for {key_i} and {key_j}. "
                      "Adjusted Rand Index might not be directly comparable. "
                      "Ensure 'participating_models' is consistent across comparisons for best results.")
                ari_score = np.nan # Indicate non-comparable
            else:
                ari_score = adjusted_rand_score(labels_i, labels_j)
            
            similarity_matrix[i, j] = ari_score

    # Plotting the heatmap
    plt.figure(figsize=(len(heatmap_labels) * 1.5, len(heatmap_labels) * 1.2)) # Dynamic sizing
    sns.heatmap(similarity_matrix, vmin = -1.0, vmax = 1.0, annot=True, cmap='viridis', fmt=".2f",
                xticklabels=heatmap_labels, yticklabels=heatmap_labels,
                cbar_kws={'label': 'Adjusted Rand Index (ARI) Similarity'},
                linewidths=.5, linecolor='black')
    plt.title('Adjusted Rand Index Similarity Between Experiment Clusterings')
    plt.xlabel('Experiment Clustering')
    plt.ylabel('Experiment Clustering')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()

    # Save the heatmap
    heatmap_filename = os.path.join(comparison_output_dir, f"{expx_id}_vs_{expy_id}_Cluster_Similarity_Heatmap_{variable_name_in_netcdf}.png")
    plt.savefig(heatmap_filename, dpi=300)
    plt.show()

print("\n--- Processing Complete ---")
```

```python
# ---- USE THIS CODE ONLY FOR TIERED EXPERIMENTS ----
# ---- Or any experiment with different number of participating models ----

# --- USER CONFIGURATION ---
# Define the actual experiment IDs for your two experiments.
# These IDs will be used in folder and file names for clarity.
expx_id = "exp08" # <--- Set the ID for your first experiment (e.g., "exp05")
expy_id = "exp10" # <--- Set the ID for your second experiment (e.g., "exp06")

# Define the paths to your experiment folders containing NetCDF files
expx_folder_path = r"C:\Users\benra\lithk_exp08" # <--- IMPORTANT: SET YOUR PATH HERE
expy_folder_path = r"C:\Users\benra\lithk_exp10" # <--- IMPORTANT: SET YOUR PATH HERE

# Define the variable name exactly as it appears in your NetCDF files
variable_name_in_netcdf = "lithk" # <--- IMPORTANT: SET YOUR VARIABLE NAME HERE

# PCA and K-Means parameters
n_components = 3 # Number of PCA components
n_clusters = 5 # Number of K-Means clusters

# Define time windows (decades as suggested)
time_windows = {
    '2015-2025': (0, 10),   # indices for 2015-2025
    '2015-2035': (0, 20),   # indices for 2015-2035  
    '2015-2045': (0, 30),   # indices for 2015-2045
    '2015-2055': (0, 40),   # indices for 2015-2055
    '2015-2065': (0, 50),   # indices for 2015-2065
    '2015-2075': (0, 60),   # indices for 2015-2075
    '2015-2085': (0, 70),   # indices for 2015-2085
    '2015-2100': (0, 85)    # full time series
}
# --- END USER CONFIGURATION ---


# --- Function to load NetCDF files from a folder ---
def load_netcdf_files_from_folder(folder_path, var_name_in_nc):
    """
    Loads NetCDF files from a specified folder, extracts a specific variable,
    and returns a list of xarray DataArrays and corresponding model names.
    """
    var_list_loaded = []
    model_names_loaded = []

    if not os.path.isdir(folder_path):
        print(f"Error: Folder not found at '{folder_path}'. Please check the path.")
        return [], []

    # Iterate through files in the folder (sorted for consistent model order)
    for filename in sorted(os.listdir(folder_path)):
        if filename.endswith(".nc"):
            filepath = os.path.join(folder_path, filename)
            try:
                with xr.open_dataset(filepath) as ds:
                    if var_name_in_nc in ds:
                        var_list_loaded.append(ds[var_name_in_nc])
                        
                        # Extract the base filename without extension
                        filename_base = os.path.splitext(filename)[0]
                        # Remove the '_expX' suffix (e.g., '_exp05' or '_exp06')
                        # and then remove the 'lithk_GIS_' prefix to get the clean model name.
                        parts = filename_base.rsplit('_', 1) # Splits at the last underscore
                        if len(parts) > 1: # Ensure there was an underscore to split by
                            model_name_with_prefix = parts[0]
                            clean_model_name = model_name_with_prefix.replace("lithk_GIS_", "")
                            model_names_loaded.append(clean_model_name)
                        else:
                            # Fallback if filename format is unexpected (e.g., no _expid)
                            model_names_loaded.append(filename_base.replace("lithk_GIS_", ""))
                        
                    else:
                        print(f"Warning: Variable '{var_name_in_nc}' not found in '{filename}'. Skipping.")
            except Exception as e:
                print(f"Error loading '{filepath}': {e}")
    return var_list_loaded, model_names_loaded

# Initialize a dictionary to store all clustering results.
# This will hold cluster labels for each experiment and time window.
all_clustering_results = {}

# Map actual experiment IDs to their folder paths
experiment_configs = {
    expx_id: expx_folder_path,
    expy_id: expy_folder_path
}

# --- Main Processing Loop for Each Experiment ---
for current_exp_id, folder_path in experiment_configs.items():
    print(f"\n--- Processing Experiment: {current_exp_id} from {folder_path} ---")

    # Load data for the current experiment
    var_list, participating_models = load_netcdf_files_from_folder(
        folder_path, variable_name_in_netcdf
    )

    if not var_list:
        print(f"No valid NetCDF files found or variable '{variable_name_in_netcdf}' missing in '{folder_path}'. Skipping {current_exp_id}.")
        continue

    # Set the variable name for plotting purposes
    var_name = variable_name_in_netcdf

    # Define base output directory for the current experiment's plots
    base_output_dir = os.path.join("Research_Plots", f"{current_exp_id}_{var_name}_plots")
    os.makedirs(base_output_dir, exist_ok=True)

    # Loop over all defined time windows
    for time_window_label, (start_idx, end_idx) in time_windows.items():
        print(f"\n Processing: {time_window_label} for {current_exp_id}...")
        
        # Create directory for each time window's plots
        folder_name = time_window_label.replace(" ", "_").replace("(","").replace(")", "")
        time_window_output_dir = os.path.join(base_output_dir, folder_name)
        os.makedirs(time_window_output_dir, exist_ok=True)

        # Get grid coordinates from the first model in var_list
        # This assumes all models share the same grid structure.
        first_var_dataarray = var_list[0]
        x_coords = first_var_dataarray['x'].values if 'x' in first_var_dataarray.coords else first_var_dataarray.coords['x'].values
        y_coords = first_var_dataarray['y'].values if 'y' in first_var_dataarray.coords else first_var_dataarray.coords['y'].values
        
        rows = len(y_coords)
        cols = len(x_coords)
        matrix_size = rows * cols

        # Step 1: Compute variable deltas for the current time window and remove NaNs
        current_window_deltas = []
        
        for var_model_data in var_list:
            # Calculate delta for the current time window using index-based selection
            delta = var_model_data.isel(time=end_idx) - var_model_data.isel(time=start_idx)
            
            # Ensure delta is resized to the common grid shape if necessary, then flatten
            delta_array = delta.values if hasattr(delta, 'values') else delta
            resized_delta = resize(delta_array, (rows, cols), anti_aliasing=True)
            flat_delta = resized_delta.flatten()
            
            current_window_deltas.append(flat_delta)

        all_deltas = np.stack(current_window_deltas)
        mask = ~np.any(np.isnan(all_deltas), axis=0) # Mask out grid points with NaN in any model
        data_matrix = all_deltas[:, mask]  # Final clean matrix (n_models, n_valid_points)
        
        # Get valid grid indices based on the mask
        gd = np.where(mask)[0]

        # Step 2: PCA (describes variability across models for this time window)
        scaler = StandardScaler()
        data_matrix_scaled = scaler.fit_transform(data_matrix)
        
        pca = PCA(n_components=n_components)
        pca_scores = pca.fit_transform(data_matrix_scaled)
        
        correct_pca_components = pca.components_ # Principal components (eigenvectors)

        # Step 3: K-Means Clustering
        kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init='auto') 
        labels = kmeans.fit_predict(pca_scores)
        
        # Store the clustering results for later comparison heatmap
        key = (current_exp_id, time_window_label)
        all_clustering_results[key] = {
            'labels': labels,
            'models': participating_models # Store models to ensure consistency during ARI calculation
        }

        # Create a DataFrame to store model names and their cluster labels
        cluster_results_df = pd.DataFrame({
            'Model': participating_models,
            'Cluster_Label': labels
        })

        # --- Cluster Color and Marker Setup for Plots ---
        if n_clusters <= 9:
            cluster_colors = [cm.get_cmap('Set1')(i) for i in range(n_clusters)]
        elif n_clusters <= 12:
            cluster_colors = [cm.get_cmap('Set3')(i) for i in range(n_clusters)]
        else:
            print("Warning: Number of clusters is > 12. Color distinctness may be challenging.")
            cluster_colors = [cm.get_cmap('hsv')(i/n_clusters) for i in range(n_clusters)]

        markers = itertools.cycle(('o', '^', 's', 'D', 'v', 'P', 'X', '*', 'H', '<', '>'))
        model_to_marker = {model: next(markers) for model in participating_models}

        # --- Step 4a: Plot PCA result in 2D (PC1 vs PC2) with Ellipses ---
        fig_2d, ax_2d = plt.subplots(figsize=(10, 8))

        for i, model_name in enumerate(participating_models): 
            cluster_id = labels[i]
            color = cluster_colors[cluster_id]
            marker = model_to_marker[model_name]

            ax_2d.scatter(
                pca_scores[i, 0], pca_scores[i, 1],
                c=[color], marker=marker, s=150, edgecolor='black', alpha=0.8
            )
        
        # Add Ellipses for each Cluster in 2D Plot
        confidence_level = 0.95
        chi2_val = stats.chi2.ppf(confidence_level, df=2) # For 2D (2 degrees of freedom)
        
        for cluster_id in np.unique(labels):
            cluster_points = pca_scores[labels == cluster_id, :2]
            if len(cluster_points) < 2: continue 

            centroid = np.mean(cluster_points, axis=0)
            covariance = np.cov(cluster_points, rowvar=False)

            eigenvalues, eigenvectors = np.linalg.eigh(covariance)
            order = eigenvalues.argsort()[::-1] # Sort for major/minor axes
            eigenvalues = eigenvalues[order]
            eigenvectors = eigenvectors[:, order]

            angle = np.degrees(np.arctan2(*eigenvectors[:, 0][::-1]))
            width = 2 * np.sqrt(eigenvalues[0] * chi2_val)
            height = 2 * np.sqrt(eigenvalues[1] * chi2_val)

            ellipse_color = cluster_colors[cluster_id]
            ellipse = Ellipse(xy=centroid, width=width, height=height, 
                              angle=angle, edgecolor=ellipse_color, fc='None', 
                              lw=2, alpha=0.7, zorder=0)
            ax_2d.add_patch(ellipse)
            
        # Create custom legend for clusters and models
        cluster_legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=cluster_colors[i], markersize=10, label=f'Cluster {i}') for i in range(n_clusters)]
        model_legend_handles = [plt.Line2D([0], [0], marker=model_to_marker[model_name], color='w', markerfacecolor=cluster_colors[labels[i]], markersize=10, label=model_name, markeredgecolor='black') for i, model_name in enumerate(participating_models)]

        first_legend = ax_2d.legend(handles=cluster_legend_handles, loc='upper left', title='Clusters', bbox_to_anchor=(1.05, 1))
        ax_2d.add_artist(first_legend)
        ax_2d.legend(handles=model_legend_handles, loc='lower left', title='Models', bbox_to_anchor=(1.05, 0))

        ax_2d.set_xlabel("PC1")
        ax_2d.set_ylabel("PC2")
        ax_2d.set_title(f"2D PCA + KMeans Clustering of {var_name} ({time_window_label})")
        plt.tight_layout(rect=[0, 0, 0.85, 1])
        plt.close()

        # --- Step 4b: Plot PCA result in 3D with Transparent Centroid Markers ---
        fig_3d = plt.figure(figsize=(12, 8))
        ax_3d = fig_3d.add_subplot(111, projection='3d')

        for i, model_name in enumerate(participating_models): 
            cluster_id = labels[i]
            color = cluster_colors[cluster_id]
            marker = model_to_marker[model_name]

            ax_3d.scatter(
                pca_scores[i, 0], pca_scores[i, 1], pca_scores[i, 2],
                c=[color], marker=marker, s=150, edgecolor='black', alpha=0.8
            )

        # Add large, transparent markers at cluster centroids in 3D
        for cluster_id in np.unique(labels):
            cluster_points_3d = pca_scores[labels == cluster_id, :]
            if len(cluster_points_3d) > 0:
                centroid_3d = np.mean(cluster_points_3d, axis=0)
                ax_3d.scatter(
                    centroid_3d[0], centroid_3d[1], centroid_3d[2],
                    c=[cluster_colors[cluster_id]], marker='o', s=1000,
                    alpha=0.1, edgecolor=cluster_colors[cluster_id], linewidth=2, zorder=0
                )

        cluster_legend_handles_3d = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=cluster_colors[i], markersize=10, label=f'Cluster {i}') for i in range(n_clusters)]
        model_legend_handles_3d = [plt.Line2D([0], [0], marker=model_to_marker[model_name], color='w', markerfacecolor=cluster_colors[labels[i]], markersize=10, label=model_name, markeredgecolor='black') for i, model_name in enumerate(participating_models)]

        first_legend_3d = ax_3d.legend(handles=cluster_legend_handles_3d, loc='center left', title='Clusters', bbox_to_anchor=(1.05, 0.7))
        ax_3d.add_artist(first_legend_3d)
        ax_3d.legend(handles=model_legend_handles_3d, loc='center left', title='Models', bbox_to_anchor=(1.05, 0.2))

        ax_3d.set_xlabel("PC1")
        ax_3d.set_ylabel("PC2")
        ax_3d.set_zlabel("PC3")
        ax_3d.set_title(f"3D PCA + KMeans Clustering of {var_name} ({time_window_label})")
        plt.tight_layout(rect=[0, 0, 0.85, 1])
        plt.close()
        
        # Step 5: Plot PCs onto the grid
        for pc_idx in range(n_components):
            pc_values = correct_pca_components[pc_idx, :] 
            
            C = np.full(matrix_size, np.nan)
            C[gd.astype(int)] = pc_values
            pc_map = C.reshape((rows, cols)) 
            
            plt.figure(figsize=(12,8))
            vmax = 0.005 # Adjust as needed for your data visualization
            
            im = plt.imshow(pc_map, cmap='bwr_r', origin='lower', vmin=-vmax, vmax=vmax)
            plt.colorbar(im, label=f'PC{pc_idx + 1}')
            plt.xlabel('x')
            plt.ylabel('y')
            plt.title(f'PC #{pc_idx + 1} ({time_window_label}) - Explained Variance: {pca.explained_variance_ratio_[pc_idx]*100:.1f}%')
            plt.axis('equal')
            plt.close()
        
        # Step 6: Plot signed contribution of each PC onto the grid
        for pc_idx in range(n_components):
            pc_values = correct_pca_components[pc_idx, :] 
            
            sum_sq_pc = np.sum(pc_values ** 2)
            if sum_sq_pc == 0:
                signed_contrib = np.zeros_like(pc_values)
            else:
                signed_contrib = 100 * np.sign(pc_values) * (pc_values ** 2) / sum_sq_pc
            
            C = np.full(matrix_size, np.nan)
            C[gd.astype(int)] = signed_contrib
            sc_map = C.reshape((rows, cols))
            
            plt.figure(figsize=(12,8))
            vmax = 0.005 # Adjust as needed for your data visualization
            
            im = plt.imshow(sc_map, cmap='bwr_r', origin='lower', vmin=-vmax, vmax=vmax)
            plt.colorbar(im, label='Signed Contribution %')
            plt.xlabel('x')
            plt.ylabel('y')
            plt.title(f'Grid contribution to PC #{pc_idx + 1} (%) - Explained Variance: {pca.explained_variance_ratio_[pc_idx]*100:.1f}% ({time_window_label})')
            plt.axis('equal')
            plt.close()
        
        # --- Step 7: Plotting the Sum of PC Vectors onto the grid ---
        summed_pc_values = np.sum(correct_pca_components, axis=0)
        ex_var = pca.explained_variance_ratio_
        cumulative_variance_of_first_four = np.cumsum(ex_var)[min(3, n_components -1)] # Get for first 4 or max available
        
        C_summed = np.full(matrix_size, np.nan)
        C_summed[gd.astype(int)] = summed_pc_values
        summed_pc_map = C_summed.reshape((rows, cols))

        plt.figure(figsize=(12,8))
        vmax_summed = np.percentile(np.abs(summed_pc_values[~np.isnan(summed_pc_values)]), 99) * 0.5
        if vmax_summed == 0: vmax_summed = 0.01

        im = plt.imshow(summed_pc_map, cmap='bwr_r', origin='lower', vmin=-vmax_summed, vmax=vmax_summed)
        plt.colorbar(im, label='Sum of All Principal Components')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title(f'Sum of All Principal Components on Grid - Cum. Variance: {cumulative_variance_of_first_four*100:.1f}% ({time_window_label})')
        plt.axis('equal')
        plt.close()

# --- Cluster Similarity Heatmap Generation ---
print("\n--- Generating Cluster Similarity Heatmap ---")

# Ensure a dedicated output directory for comparison plots
comparison_output_dir = os.path.join("Research_Plots", "Cluster_Comparisons")
os.makedirs(comparison_output_dir, exist_ok=True)

# Get sorted keys for consistent ordering in the heatmap (e.g., ('exp05', '2015-2100'), ('exp06', '2015-2100'))
comparison_keys = sorted(list(all_clustering_results.keys()))

if len(comparison_keys) < 2:
    print("Not enough clustering results to generate a similarity heatmap (need at least two experiments/time windows).")
else:
    # Create labels for the heatmap axes (e.g., "exp05_2015-2100")
    heatmap_labels = [f"{exp}_{tw}" for exp, tw in comparison_keys]

    # Initialize similarity matrix
    num_comparisons = len(comparison_keys)
    similarity_matrix = np.zeros((num_comparisons, num_comparisons))

    # Calculate pairwise Adjusted Rand Index (ARI) scores
    for i in range(num_comparisons):
        for j in range(num_comparisons):
            key_i = comparison_keys[i]
            key_j = comparison_keys[j]

            labels_i = all_clustering_results[key_i]['labels']
            models_i = all_clustering_results[key_i]['models']

            labels_j = all_clustering_results[key_j]['labels']
            models_j = all_clustering_results[key_j]['models']

            # --- MODIFIED SECTION FOR HANDLING DIFFERING MODEL LISTS ---
            # Find common models between the two lists
            common_models = sorted(list(set(models_i).intersection(set(models_j))))

            if not common_models:
                print(f"Warning: No common models found for {key_i} and {key_j}. Setting ARI to NaN.")
                ari_score = np.nan
            elif len(common_models) < 2:
                # ARI is not meaningful for fewer than 2 common elements
                print(f"Warning: Fewer than 2 common models for {key_i} and {key_j}. Setting ARI to NaN.")
                ari_score = np.nan
            else:
                # Create dictionaries to map model names to their labels
                model_to_label_i = dict(zip(models_i, labels_i))
                model_to_label_j = dict(zip(models_j, labels_j))

                # Extract labels for only the common models, ensuring consistent order
                labels_for_common_i = np.array([model_to_label_i[model] for model in common_models])
                labels_for_common_j = np.array([model_to_label_j[model] for model in common_models])

                ari_score = adjusted_rand_score(labels_for_common_i, labels_for_common_j)
            
            similarity_matrix[i, j] = ari_score

    # Plotting the heatmap
    plt.figure(figsize=(len(heatmap_labels) * 1.5, len(heatmap_labels) * 1.2)) # Dynamic sizing
    sns.heatmap(similarity_matrix, vmin = -1.0, vmax = 1.0, annot=True, cmap='viridis', fmt=".2f",
                xticklabels=heatmap_labels, yticklabels=heatmap_labels,
                cbar_kws={'label': 'Adjusted Rand Index (ARI) Similarity'},
                linewidths=.5, linecolor='black')
    plt.title('Adjusted Rand Index Similarity Between Experiment Clusterings')
    plt.xlabel('Experiment Clustering')
    plt.ylabel('Experiment Clustering')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()

    # Save the heatmap
    heatmap_filename = os.path.join(comparison_output_dir, f"{expx_id}_vs_{expy_id}_Cluster_Similarity_Heatmap_{variable_name_in_netcdf}.png")
    plt.savefig(heatmap_filename, dpi=300)
    plt.show()

print("\n--- Processing Complete ---")
```
